```python
# Sigmoid

import numpy as np
import matplotlib.pylab as plt

def sigmoid_function(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5, 5)
y = sigmoid_function(x) # 0과 1사이를 왔다갔다하는 함수

plt.plot(x, y)
plt.show()
```

```python
#tanh

def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5)
y = tanh_function(x)

plt.plot(x, y)
plt.show()
```

```python
# ReLU(Rectified Linear Unit)

def relu_function(x):
    return np.where(x <= 0, 0, x)
x = np.linspace(-5, 5)
y = relu_function(x)
plt.plot(x, y)
plt.show()
```

```python
# Leaky ReLU

def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x)
x = np.linspace(-5, 5)
y = leaky_relu_function(x)
plt.plot(x, y)
plt.show()
```

```python
def softmax_function(x):
    return np.exp(x)/np.sum(np.exp(x)) # 소프트맥스함수
y = softmax_function(np.array([1,2,3]))
print(y)
```

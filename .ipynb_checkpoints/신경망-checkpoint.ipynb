신경세포(=뉴런) 뇌를 이루어짐.


Activation function 활성화 함수
출력은 u라고한다면 f(u) 최종 결과는 y

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(-1.0, 1.0, 0.2)
x2 = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10, 10))

#w_x1 = 2.5
#w_x2 = 3.0 값을 고정
W = np.array([2.5, 3.0])

#bias = 0.1
bias = np.array([0.1])
```

```python
for i in range(10):
    for j in range(10):
        X = np.array([x1[i], x2[j]])
        #u = x1[i] * w_x1 + x2[j] * w_x2 + bias # 이해하기
        u = np.dot(X, W.T) + bias  # W.T 세워주는 것 행렬의 곱셈 사용
        y = 1 / (1 + np.exp(-u))
        Z[j][i] = y
```

```python
plt.imshow(Z, "gray", vmin = 0.0, vmax = 1.0)
plt.colorbar()
plt.show()
```
